# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame creation (you should use your actual DataFrame here)
import pandas as pd
import numpy as np
# Creating a sample DataFrame
np.random.seed(0)
df = pd.DataFrame({
    'num1': np.random.randn(100),
    'num2': np.random.randn(100) * 10,
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})
# List of numerical columns (you should use your actual numerical columns)
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# Histograms
plt.figure(figsize=(12, 6))
n_cols = min(2, len(numerical_cols))  # Number of columns for subplots
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols  # Calculate the number of rows needed
for i, col in enumerate(numerical_cols):
    plt.subplot(n_rows, n_cols, i + 1)  # Adjust the subplot index
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_cols])
plt.title('Boxplot of Numerical Columns')
plt.xticks(rotation=45)
plt.show()
# Bar chart for categorical column
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
if 'Category' in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Category', data=df)
    plt.title('Bar Chart of Category')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Category column not found in DataFrame.")
```

# Standardization of Numerical Variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
# Example DataFrame creation (replace this with your actual DataFrame)
np.random.seed(0)
df = pd.DataFrame({
    'num1': np.random.randn(100),
    'num2': np.random.randn(100) * 10
})
# List of numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# Standardization (z-score normalization)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_cols])
scaled_df = pd.DataFrame(scaled_data, columns=numerical_cols)
# Comparison of data distributions before and after standardization
plt.figure(figsize=(12, 6))
n_cols = len(numerical_cols)  # Number of columns for subplots
for i, col in enumerate(numerical_cols):
    plt.subplot(1, n_cols, i + 1)  # Adjust the subplot index
    sns.histplot(df[col], kde=True, color='blue', label='Original', stat='density')
    sns.histplot(scaled_df[col], kde=True, color='red', label='Standardized', stat='density')
    plt.legend()
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# One-hot encoding
encoded_df = pd.get_dummies(df, columns=numerical_cols, drop_first=True)
# Display a portion of the transformed dataset
encoded_df.head()
```
